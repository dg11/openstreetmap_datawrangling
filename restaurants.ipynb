```python
import csv
```

```python
map = {
    'Asian_Fusion': 'asian',
    'Argentinian': 'argentinian',
    '*': 'unknown',
    'American, home cooking': 'american',
    'BBQ': 'barbecue',
    'Bar-B-Q': 'barbecue',
    'Coffee_shop': 'coffee_shop',
    'Coffee_shop,_Bar': 'coffee_shop',
    'Cuban': 'cuban',
    'Donuts,_Breakfast,_Ice_Cream,_Treats': 'donut',
    'doughnut': 'donut',
    'donuts': 'donut',
    'Frozen_Yogurt': 'frozen_yogurt',
    'Greek-centric_Mediterranean_cuisine': 'greek/mediterranean',
    'Greek/Mediteranian': 'greek/mediterranean',
    'Hamburgers and fires': 'burger',
    'ICe_cream':  'ice_cream',
    'Ice_cream':  'ice_cream',
    'Jamaican,_Cuban': 'jamaican',
    'Mexican_Korean': 'mexican/korean',
    'Moroccan': 'moroccan',
    'Sandwhich shop': 'sandwich',
    'Surf_&_Turf': 'surf_and_turf',
    'Tacos,_coffee': 'coffee_shop',
    'Tex-Mex': 'tex_mex',
    'tex-mex': 'tex_mex',
    'Texas Style BBQ': 'barbecue',
    'american,wings': 'american',
    'burger;sandwich;hotdog;grill;irish;fish_and_chips;vegetarian;mexican': 'burger'
        }
```

```python
with open('atx_restaurants.csv', 'r') as f:
    reader = csv.DictReader(f)
    atxres = []
    for line in reader:
        data = {}
        data['nodes_lat'] = line['nodes_lat']
        data['nodes_lon'] = line['nodes_lon']
        if line['value'] in map.keys():
            data['value'] = map[line['value']]
        else:
            data['value'] = line['value']
        data['id'] = line['id']
        atxres.append(data)
```

```python
import pandas as pd
```

```python
atxres_df = pd.DataFrame.from_dict(atxres)
```
